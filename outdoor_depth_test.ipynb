```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
depth_npy_path = "/scratch2/isidora/pano-infinigen/outputs/my_outdoor_dataset2/11de797a/frames/Depth/camera_0/Depth_0_0_0048_0.npy"
```

```python
img_arr = np.load(depth_npy_path)
print(img_arr.shape)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
# plot values
plt.hist(img_arr.ravel(), bins=100)
plt.title("Histogram of Depth Values")
plt.xlabel("Depth Value")
plt.ylabel("Frequency")
plt.xlim(0, 1000)  # Adjust x-axis limit as needed
plt.grid()
```

```python
# log of depth image
img_arr_log = np.log(img_arr + 1e-6)  # Adding a small constant to avoid log(0)
plt.imshow(img_arr_log)
plt.title("Log Depth Image")
plt.show()
```

```python
# clip values to 200
img_arr_clipped = np.clip(img_arr_log, 0, 75)
plt.imshow(img_arr_clipped)
plt.title("Clipped and Log Depth Image")
plt.show()
```

```python
# just clip
img_arr_clipped = np.clip(img_arr, 0, 75)
plt.imshow(img_arr_clipped)
plt.title("Clipped Depth Image")
plt.show()
```

```python
# plot histogram of clipped values
plt.hist(img_arr_clipped.ravel(), bins=100)
plt.title("Histogram of Clipped Depth Values")
plt.xlabel("Clipped Depth Value")
plt.ylabel("Frequency")
plt.xlim(0, 200)  # Adjust x-axis limit as needed
plt.grid()
plt.show()
```
